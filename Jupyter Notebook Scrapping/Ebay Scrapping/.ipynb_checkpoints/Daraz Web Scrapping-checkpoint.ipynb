```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd  # Import pandas for DataFrame

# Set up Chrome options
options = Options()
options.add_argument('--headless')  # Run in headless mode (no UI)
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Set up the WebDriver (ensure you have the correct WebDriver installed)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# URL of the page to scrape
my_url = "https://www.daraz.pk/catalog/?spm=a2a0e.tm80335142.search.2.62aa7bdalagPsl&q=mobile%20phone&_keyori=ss&clickTrackInfo=textId--5704130092250749346__abId--None__pvid--72963c26-965a-4a28-922a-67d93cf8f23e__matchType--1__abGroup--None__srcQuery--mobile%20phone__spellQuery--mobile%20phone__ntType--nt-common&from=suggest_normal&sugg=mobile%20phone_0_1"

# Navigate to the page
driver.get(my_url)

# Wait for the page to load
time.sleep(5)  # Adjust this time as needed for your connection

# Find the container with class '_17mcb'
container = driver.find_element(By.CLASS_NAME, '_17mcb')

# Find all product items within the container
items = container.find_elements(By.CLASS_NAME, 'Bm3ON')

# Initialize product data list
product_data = []

# Loop through each product item
for item in items:
    # Extract product name
    name = item.find_element(By.CSS_SELECTOR, 'a[title]').text.strip()
    
    # Extract price
    price = item.find_element(By.CLASS_NAME, 'ooOxS').text.strip()
    
    # Extract image URL
    image_url = item.find_element(By.TAG_NAME, 'img').get_attribute('src')
    
    # Extract number sold and location
    sold = item.find_element(By.CLASS_NAME, '_1cEkb').text.strip()
    location = item.find_element(By.CLASS_NAME, 'oa6ri').get_attribute('title')

    # Append extracted data to list
    product_data.append({
        'Name': name,
        'Price': price,
        'Image URL': image_url,
        'Sold': sold,
        'Location': location
    })

# Close the browser
driver.quit()

# Create a Pandas DataFrame from the product data
df = pd.DataFrame(product_data)

# Display the DataFrame
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Set up Chrome options
options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Initialize the WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# URL to scrape
url = "https://www.daraz.pk/products/chia-seeds-150-grams-organic-and-original-superfood-herbal-diet-supplement-for-weight-loss-i418697553-s2082506387.html"

# Open the URL
driver.get(url)

# Wait for the page to load and the reviews section to be present
try:
    # Adjust the timeout as necessary
    WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, 'item')))
except Exception as e:
    print(f"Error while waiting for the page to load: {e}")
    driver.quit()

# Parse the page source
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Print the page source for debugging
# Uncomment the line below to see the full HTML structure
# print(soup.prettify())

# Find all reviews
reviews = soup.find_all('div', class_='item')

# Check if there are any reviews
if not reviews:
    print("No reviews found.")
else:
    # Extract and print review data
    for review in reviews:
        try:
            # Review title and date
            date_elem = review.find('span', class_='title right')
            date = date_elem.text.strip() if date_elem else "No date found"
            
            # Reviewer username
            username_elem = review.find('span', class_='middle')
            username = username_elem.find_all('span')[0].text.strip() if username_elem and username_elem.find_all('span') else "No username found"
            
            # Review content
            content_elem = review.find('div', class_='content')
            content = content_elem.text.strip() if content_elem else "No content found"
            
            # Review images
            images = review.find_all('div', class_='review-image__item')
            image_links = [img['style'].split('url("')[1].split('")')[0] for img in images if 'style' in img.attrs]

            # Print the extracted data
            print("\n--- Review ---")
            print(f"Date: {date}")
            print(f"Username: {username}")
            print(f"Review: {content}")
            if image_links:
                print(f"Images: {', '.join(image_links)}")
            else:
                print("Images: No images provided")
            print("=" * 50)  # Separator for each review
            
        except Exception as e:
            print(f"Error extracting data: {e}")

# Close the WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Set up Chrome options
options = Options()
options.add_argument("--headless")  # Run in headless mode for background execution
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Initialize the WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# URL to scrape
url = "https://www.daraz.pk/products/chia-seeds-150-grams-organic-and-original-superfood-herbal-diet-supplement-for-weight-loss-i418697553-s2082506387.html?pvid=7c89e782-de9d-42b1-bead-b9385f2c4874&search=jfy&scm=1007.28811.410576.0&priceCompare=skuId%3A2082506387%3Bsource%3Atpp-recommend-plugin-41701%3Bsn%3A7c89e782-de9d-42b1-bead-b9385f2c4874%3BunionTrace%3A2101066517302741638428190e1789%3BoriginPrice%3A33000%3BvoucherPrice%3A33000%3BdisplayPrice%3A33000%3BsourceTag%3A%23auto_collect%231%24auto_collect%24%3BsinglePromotionId%3A-1%3BsingleToolCode%3A-1%3BvoucherPricePlugin%3A1%3BbuyerId%3A0%3ButdId%3A-1%3Btimestamp%3A1730274163926&spm=a2a0e.tm80335142.just4u.d_418697553"

# Open the URL
driver.get(url)

# Wait for the page to fully load
time.sleep(10)

all_reviews = []  # List to hold all review data

while True:
    # Parse the page source
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Find all reviews on the current page
    reviews = soup.find_all('div', class_='item')

    # Check if there are any reviews
    if not reviews:
        print("No reviews found on this page.")
    else:
        # Extract review data
        for review in reviews:
            try:
                # Review title and date
                date = review.find('span', class_='title right').text.strip()
                
                # Reviewer username
                username = review.find('span', class_='middle').find_all('span')[0].text.strip()
                
                # Review content
                content = review.find('div', class_='content').text.strip()
                
                # Review images
                images = review.find_all('div', class_='review-image__item')
                image_links = [img['style'].split('url("')[1].split('")')[0] for img in images]
                
                # Store the extracted data
                all_reviews.append({
                    "Date": date,
                    "Username": username,
                    "Review": content,
                    "Images": image_links
                })

            except Exception as e:
                print(f"Error extracting data: {e}")

    # Find the next pagination button
    next_button = soup.find('button', class_='next-btn next-btn-normal next-btn-medium next-pagination-item next')
    
    # If there is no next button, break the loop
    if not next_button or 'disabled' in next_button.get_attribute('outerHTML'):
        print("No more pages to scrape.")
        break
    
    # Click the next button
    driver.execute_script("arguments[0].click();", next_button)
    
    # Wait for the next page to load
    time.sleep(10)

# Print all collected reviews
for review in all_reviews:
    print("\n--- Review ---")
    print(f"Date: {review['Date']}")
    print(f"Username: {review['Username']}")
    print(f"Review: {review['Review']}")
    if review['Images']:
        print(f"Images: {', '.join(review['Images'])}")
    else:
        print("Images: No images provided")
    print("=" * 50)  # Separator for each review

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
from textblob import TextBlob
import re
import nltk
from nltk.corpus import stopwords

# Ensure you have stopwords for NLP processing
nltk.download('stopwords')

# Load your data (assuming 'filtered_df' is already defined)
# Example: filtered_df = pd.read_csv("Reviews.csv")

# Roman Urdu to English sentiment dictionary
roman_urdu_translation = {
    "boht achi": "very good",
    "zabardast": "excellent",
    "acha": "good",
    "behtareen": "excellent",
    "pasand aya": "liked it",
    "boht buri": "very bad",
    "bekar": "useless",
    "kharab": "damaged",
    "bhot bura": "very bad",
    "pasand nahi aya": "did not like"
}

# Function to clean the review text
def clean_text(text):
    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    text = " ".join([word for word in text.split() if word not in stop_words])
    return text

# Function to translate Roman Urdu phrases to English
def translate_roman_urdu(text):
    for roman, english in roman_urdu_translation.items():
        if roman in text:
            text = text.replace(roman, english)
    return text

# Custom function to perform sentiment analysis
def predict_sentiment(review_text):
    # Translate Roman Urdu phrases before analysis
    translated_text = translate_roman_urdu(review_text)
    
    # Use TextBlob for sentiment analysis
    analysis = TextBlob(translated_text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity == 0:
        return "Neutral"
    else:
        return "Negative"

# Apply text cleaning, translation, and sentiment analysis to filtered DataFrame
filtered_df['Cleaned_Review'] = filtered_df['Review'].apply(clean_text)
filtered_df['Translated_Review'] = filtered_df['Cleaned_Review'].apply(translate_roman_urdu)
filtered_df['Sentiment'] = filtered_df['Translated_Review'].apply(predict_sentiment)

# Display only rows where the 'Translated_Review' has significant content
filtered_df = filtered_df[filtered_df['Translated_Review'] != ""]

# Show a subset of columns for easier reading
display_columns = ['Date', 'Username', 'Review', 'Cleaned_Review', 'Translated_Review', 'Sentiment']
filtered_df[display_columns]
```
